```python
# Recursive implementation for finding the nth term in our example sequence S
# We're assuming the input n is an integer greater than or equal to 1
calls = 0
count = 0
one = 0
e = 0
def S(n):
    global calls
    global count
    global one
    global e
        
    calls = calls + 1
#     print("calls", calls)
    if n == 1:
        one += 1
        count += 2
        return 1
    else:
        count += 7
        e += 1
        return S(n-1) + S(n-1)

print("S(n) = ", S(4))
print("calls = ", calls)
print("count = ", count)
print("one = ", one)
print("e = ", e)
```
